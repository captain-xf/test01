## Convert

```python
codelib = '''@B%8&WM#"oahkbdpqwmZOBQLCJUYXzcwumxrjft/\|()1{}[]?-_+~<>l!l\
I;:,"^;". '''
```

```python
codelib
```

```python
vg.max(), vg.min(), xf.max(), xf.min()
```

```python
def transform(image, codelist):
    h, w = image.shape
    count = len(codelist)
    codePic = ''
    for i in range(h):
        for j in range(w):
            k = int(image[i,j]*(count-1)/float(image.max()))
            codePic = codePic + codelist[k]
        codePic += '\r'
    return codePic
```

```python
%%time
vg1 = transform(vg, codelib)
```

```python
%%time
xf1 = transform(xf, codelib)
```

```python
with open('XF.txt', 'wb') as f:
    f.write(xf1)
```
