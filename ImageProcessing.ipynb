```python
from scipy import misc
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable, ImageGrid
```

```python
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 10
```

```python
vg01 = misc.imread(r'./pictures/von_gogh_04.jpg', mode='L')
vg02 = misc.imread(r'./pictures/von_gogh_05.jpg', mode='L')
```

```python
styles = ['jet', 'viridis', 'rainbow']
# names = ['vongogh_Starry_Night_Rhone', 'vongogh_Starry_Night']
names = ['vongogh_Night_Cafe', 'vongogh_Stevedores']
for vg, name in zip([vg01, vg02], names):
    fig, axs = plt.subplots(1, 3, sharey=True)
    fig.set_size_inches(6, 3.2)
    fig.subplots_adjust(left=0.01, top=0.95, wspace=0.08)
    j = 0
    for style in styles:
        ax = axs[j]; plt.sca(ax)
        im = plt.imshow(vg, cmap=plt.cm.__dict__[style])
        plt.title(style, fontsize=9)
        ax.tick_params(axis='both', labelsize=8)
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", "5%", pad=0.05)
        cbar = plt.colorbar(im, cax=cax)
        cbar.ax.set_axis_off()
        j = j + 1
    cbar.ax.set_axis_on()
    cbar.ax.tick_params(labelsize=8)
    plt.savefig(r'./pictures/'+name, dpi=200, bbox_inches='tight', 
                transparent=True)
```
