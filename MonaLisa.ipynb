```python
m1 = ocr.split(' ')
```

```python
m2 = map(int, m1)
```

```python
m3 = array(m1, dtype=int)
```

```python
mona = m3.reshape([22, 20])
```

```python
# comparison
stys = ['jet', 'rainbow', 'viridis']
fig, axs = subplots(1, 3, sharey=True)
fig.set_size_inches(6, 3.5)
fig.subplots_adjust(left=0.01, top=0.95, wspace=0.25)
j = 0
for sty in stys:
    rcParams['image.cmap'] = sty
    ax = axs[j]; sca(ax)
    im = imshow(mona); 
    locator_params(axis='x',nbins=3)
    title(sty)
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="8%", pad=0.08)
    colorbar(im, cax=cax)
    j = j + 1
savefig('MonaLisa', dpi=200, bbox_inches='tight', transparent=True)
```
